```python
def state_heuristic(row):
    dist = row['district']
    if "'" in row['district']:
        dist = dist.strip().split("'")
    else:
        dist = dist.strip().split(" ")
    return dist[0]
    
def last_name(row):
    name = row['candidate'].strip().split(" ")
    return name[-1]
```
