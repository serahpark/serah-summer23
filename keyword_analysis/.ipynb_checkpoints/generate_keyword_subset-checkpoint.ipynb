# Subset of corpus with keyword
As a part of exploring the research question of `crime` as a dogwhistle, this notebook filters through a provided corpus, `general_candidates_corpus.csv`, and returns a new filtered corpus, `general_candidates_crime_subset.csv`, with only emails containing the given keyword, `crime`. Corpus & keywords can be interchanged.

```python
import pandas as pd
```

```python
corpus = pd.read_csv('../corpus_sample/general_candidates_corpus.csv', index_col=0)
```

```python
keyword = 'crime'
```

```python
emails_with_keyword = []

for index, row in corpus.iterrows():
    if keyword in str(row['body_text']):
        emails_with_keyword.append(row)
        
subset = pd.concat(emails_with_keyword, axis=1).T
        
subset.reset_index(drop=True, inplace=True)
```

```python
subset
```

```python
subset.to_csv('general_candidates_crime_subset.csv')
```
