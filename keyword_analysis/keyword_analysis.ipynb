# Initial Keyword Analysis
First aggregation by keyword & visualization of these distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)
import seaborn as sns
sns.set_theme()
```

```python
candidate_info = pd.read_csv('aggregated_gen_candidates.csv', index_col=0)
candidates_list = candidate_info['name'].to_list()
corpus = pd.read_csv('../corpus_sample/primary_candidates_corpus.csv', usecols=['name', 'office_sought', 'party_affiliation', 'office_level', 'district_type', 'body_text', 'uid_email'])
```

```python
def count_keyword(name):
    candidate_emails = corpus.loc[corpus['name'] == name]
    keyword_count = 0
    for index, row in candidate_emails.iterrows():
        if keyword in str(row['body_text']):
            keyword_count += 1
            print(index)
    return keyword_count

keyword = "lynching"
candidate_info['keyword_count'] = candidate_info['name'].apply(count_keyword)
candidate_info['keyword_proportion'] = candidate_info['keyword_count'] / candidate_info['total_emails_sent']
```

```python
candidate_info.sort_values('keyword_proportion', ascending=False)
# try splitting by both opponent's race AND candidate's political affiliation and print out # of candidates
```

```python
def unique_senders(w):
    senders = set()
    for index, row in corpus.iterrows():
        if w in str(row['body_text']):
            senders.add(row['name'])
    return len(senders)

unique_senders('crime')
```
