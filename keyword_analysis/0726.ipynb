I took all of the candidates in the corpus running for federal positions and joined them with data on their opponent as well as the information I collected about Black candidates to identify whether a candidate's opponent is Black. Here, we also indicate the number of total emails sent by the candidate.

Note that there is still an error to be fixed in data wrangling - some Senate candidates listed here did not make it to the general election, but I haven't gotten to filtering them out. However, no Senate candidates listed here have Black opponents.

```python
# for one keyword
keyword = "crime"
def count_keyword(name):
    candidate_emails = corpus.loc[corpus['name'] == name]
    keyword_count = 0
    for index, row in candidate_emails.iterrows():
        if keyword in str(row['body_text']):
            keyword_count += 1
#             print(index)
    return keyword_count

# # for a set of keywords
# keywords = ["tough on crime", "law and order", "clean up our streets", "cleaning up our streets", "urban", "security from unrest", "war on crime", "law & order"]
# def count_keywords(name):
#     candidate_emails = corpus.loc[corpus['name'] == name]
#     track_keyword = []
#     for index, row in candidate_emails.iterrows():
#         track_keyword.append(any(keyword in str(row['body_text']) for keyword in keywords))
#     return sum(track_keyword)

# # for a set of keywords
# keywords = ["tough on crime", "law and order", "clean up our streets", "cleaning up our streets"]
# def many_keywords(name):
#     candidate_emails = corpus.loc[corpus['name'] == name]
#     track_keyword = []
#     key_count = 0
#     for index, row in candidate_emails.iterrows():
#         for word in keywords:
#             if word in str(row['body_text']):
#                 key_count += 1
#         track_keyword.append(key_count > 2)
#         if key_count > 1:
#             print(index)
#         return sum(track_keyword)
```

## Keyword Distributions

The data used here is all emails sent by candidates in general (except Senate) elections, including emails with and without "crime" - hence why most of these distributions have peaks at 0.

Visualizing the distribution of proportions as a whole:

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

### Distribution by count/frequency at proportion

```python
sns.displot(data=candidate_info, x='keyword_proportion').set(title="Frequency of 'crime' mentions")#, kind="kde")#, hue="party_affiliation")
# sns.displot(data=candidate_info, x='keyword_proportion', kde=True)
```

Calculating the mean proportions of mentions of 'crime' across candidates with Black and non-Black opponents:

```python
keyword_vs_race = candidate_info.groupby('is_black')['keyword_proportion'].mean()
keyword_vs_race
```

```python
plt.bar(range(len(keyword_vs_race)), keyword_vs_race.values)
plt.xlabel("Candidate's Opponent is Black")
plt.ylabel("Average Proportions")
plt.title("Average Mentions of 'Crime'")

plt.xticks(range(len(keyword_vs_race)), keyword_vs_race.index)
```
