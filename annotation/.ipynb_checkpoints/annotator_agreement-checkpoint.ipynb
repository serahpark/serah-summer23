```python
import pandas as pd
```

```python
sample = pd.read_csv('email_labeling/labeling_set.csv', index_col=0)
```

```python
sample
```

```python
katie = pd.read_csv('annotated_csv/katie_annotated - katie.csv', usecols=['uid_email', 'annotator_1_A', 'annotator_1_B', 'annotator_1_C'])
mark = pd.read_csv('annotated_csv/mark_annotated - mark.csv', usecols=['uid_email', 'annotator_1_A', 'annotator_1_B', 'annotator_1_C'])
mark = mark.fillna(0)
matt = pd.read_csv('annotated_csv/matt_annotated - matt.csv', usecols=['uid_email', 'annotator_2_A', 'annotator_2_B', 'annotator_2_C'])
serah = pd.read_csv('annotated_csv/serah_annotated - serah.csv', usecols=['uid_email', 'annotator_2_A', 'annotator_2_B', 'annotator_2_C'])
serah = serah.fillna(0)
```

```python
merged_katie = sample.merge(katie, on='uid_email', how='outer')
```

```python
merged_km = merged_katie.merge(mark, on='uid_email', how='outer')
merged_kmm = merged_km.merge(matt, on='uid_email', how='outer')
merged = merged_kmm.merge(serah, on='uid_email', how='outer')
merged.head(60)
```

```python
merged.columns
```

```python
merged['annotator_1_A_x'] = merged.apply(
    lambda row: row['annotator_1_A_x'] if pd.notnull(row['annotator_1_A_x']) else row['annotator_1_A_y'],
    axis=1)

merged['annotator_1_B_x'] = merged.apply(
    lambda row: row['annotator_1_B_x'] if pd.notnull(row['annotator_1_B_x']) else row['annotator_1_B_y'],
    axis=1)

merged['annotator_1_C_x'] = merged.apply(
    lambda row: row['annotator_1_C_x'] if pd.notnull(row['annotator_1_C_x']) else row['annotator_1_C_y'],
    axis=1)

merged['annotator_2_A_x'] = merged.apply(
    lambda row: row['annotator_2_A_x'] if pd.notnull(row['annotator_2_A_x']) else row['annotator_2_A_y'],
    axis=1)

merged['annotator_2_B_x'] = merged.apply(
    lambda row: row['annotator_2_B_x'] if pd.notnull(row['annotator_2_B_x']) else row['annotator_2_B_y'],
    axis=1)

merged['annotator_2_C_x'] = merged.apply(
    lambda row: row['annotator_2_C_x'] if pd.notnull(row['annotator_2_C_x']) else row['annotator_2_C_y'],
    axis=1)

# Drop the duplicate columns
merged = merged.drop(columns=['annotator_1_A_y', 'annotator_1_B_y', 'annotator_1_C_y', 'annotator_2_A_y', 'annotator_2_B_y', 'annotator_2_C_y'])
```

```python
merged = merged.rename(columns={'annotator_1_A_x': 'annotator1A', 'annotator_1_B_x': 'annotator1B', 'annotator_1_C_x': 'annotator1C', 'annotator_2_A_x': 'annotator2A', 'annotator_2_B_x': 'annotator2B', 'annotator_2_C_x': 'annotator2C'})
```

```python
merged['agreementA'] = (merged['annotator1A'] == merged['annotator2A'])
merged['agreementB'] = (merged['annotator1B'] == merged['annotator2B'])
merged['agreementC'] = (merged['annotator1C'] == merged['annotator2C'])
```
