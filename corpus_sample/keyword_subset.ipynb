```python
import pandas as pd
```

```python
corpus = pd.read_csv('../corpus_sample/filtered_corpus.csv', usecols=['name', 'office_sought', 'party_affiliation', 'office_level', 'district_type', 'body_text', 'uid_email'])
```

```python
corpus = corpus.loc[corpus['office_level'] == 'Federal']
```

```python
keyword = 'crime'
```

```python
emails_with_keyword = []

for index, row in corpus.iterrows():
    if keyword in str(row['body_text']):
        emails_with_keyword.append(row)
        
subset = pd.concat(emails_with_keyword, axis=1).T
        
subset.reset_index(drop=True, inplace=True)
```

```python
subset
```

```python
subset.to_csv('crime_subset.csv')
```
